```python
input_data = dataset[["Age"]]
output_data = dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(input_data,output_data,train_size=0.8,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x=input("Enter your Age:- ")
lr.predict([[x]])
```

```python
lr.predict(x_test)
```

```python
lr.score(x_test,y_test),lr.score(x_train,y_train)
```

```python
plt.scatter(dataset["Age"],dataset["Purchased"])
plt.plot(dataset["Age"],lr.predict(dataset[["Age"]]),c="b")
plt.grid()
```

```python
plt.scatter(dataset["Age"],dataset["Purchased"])
plt.plot(x_test,lr.predict(x_test),c="k")
plt.grid()
```
